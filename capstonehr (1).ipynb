### Outlier Treating 

```python
x=pd.DataFrame(data=df,columns=['Bonus','Unit_Price','Months','Education'])
y=pd.DataFrame(df['Salary'])
# dependent and independent variable
```

```python
q1=np.quantile(x['Bonus'],0.25)
q2=np.quantile(x['Bonus'],0.50)
q3=np.quantile(x['Bonus'],0.75)
print(f'q1={q1} | q2={q2} |q3={q3}')
IQR=q3-q1
upperwhisker=q3+(1.5*IQR)
lowerwhisker=q1-(1.5*IQR)
x['Bonus']=x['Bonus'][(x['Bonus']>=lowerwhisker)&(x['Bonus']<=upperwhisker)]
x['Bonus'].fillna(float(x['Bonus'].median()),inplace=True)
sns.boxplot(x['Bonus'])
#handling outliers for Bonus using IQR
```

```python
x['Unit_Price']=np.log(x['Unit_Price'])
#unit price was skewed so first performed the log transfrmation
q1=np.quantile(x['Unit_Price'],0.25)
q2=np.quantile(x['Unit_Price'],0.50)
q3=np.quantile(x['Unit_Price'],0.75)
print(f'q1={q1} | q2={q2} |q3={q3}')
IQR=q3-q1
upperwhisker=q3+(1.5*IQR)
lowerwhisker=q1-(1.5*IQR)
x['Unit_Price']=x['Unit_Price'][(x['Unit_Price']>=lowerwhisker)&(x['Unit_Price']<=upperwhisker)]
x['Unit_Price'].fillna(float(x['Unit_Price'].median()),inplace=True)
sns.boxplot(x['Unit_Price'])
#handling outliers for unit price using IQR
```

```python
y['Salary']=np.log(y['Salary'])
#unit price was skewed so first performed the log transfrmation
q1=np.quantile(y['Salary'],0.25)
q2=np.quantile(y['Salary'],0.50)
q3=np.quantile(y['Salary'],0.75)
print(f'q1={q1} | q2={q2} |q3={q3}')
IQR=q3-q1
upperwhisker=q3+(1.5*IQR)
lowerwhisker=q1-(1.5*IQR)
y['Salary']=y['Salary'][(y['Salary']>=lowerwhisker)&(y['Salary']<=upperwhisker)]
y['Salary'].fillna(float(y['Salary'].median()),inplace=True)
sns.boxplot(y['Salary'])
#handling outliers for Salary using IQR
```

## Decision TreeRegressor 

```python
#Decison tree is robust to outliers and it can handle data if data is not romally distributed
from sklearn.tree import DecisionTreeRegressor #importing the regressor
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)

regressor = DecisionTreeRegressor(max_depth=4) # using max depth = 3 efficiency is around 93% but max depth = 4 higher efficiency  
regressor.fit(x_train, y_train)
#fit the regressor in x and y
y_pred=regressor.predict(x_train)
print('r2 score for testing :' ,r2_score(y_train,y_pred))
print('Mean absolute error for testing:' , metrics.mean_absolute_error(y_train,y_pred))
print('Mean absolute error for traning :' , metrics.mean_squared_error(y_train,y_pred))
print('Root mean squared error for training :' , np.sqrt(metrics.mean_squared_error(y_train,y_pred)))
print()
y_pred=regressor.predict(x_test)
print('r2 score for testing :' ,r2_score(y_test,y_pred))
print('Mean absolute error for testing:', metrics.mean_absolute_error(y_test,y_pred))
print('Mean squared testing:', metrics.mean_squared_error(y_test,y_pred))
print('Root Mean Squared error for testing:' , np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
#importing shfflesplit using sklearn libary for cross validation
from sklearn.model_selection import ShuffleSplit
model=DecisionTreeRegressor(max_depth=4)
ssplit=ShuffleSplit(n_splits=10, test_size=0.20)
from sklearn.model_selection import cross_val_score
results=cross_val_score(model,x,y,cv=ssplit)
print(results)
print(np.mean(results))
#after cross validation we got 98%.
```

## Random Forest Regressor 

```python
#applying the random forest regressor model
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.metrics import r2_score
regressor = RandomForestRegressor(n_estimators=100,random_state=0)
regressor.fit(x_train,y_train)
#fit the regressor in x and y
y_pred=regressor.predict(x_train)
print('r2 score for testing :' ,r2_score(y_train,y_pred))
print('Mean absolute error for testing:' , metrics.mean_absolute_error(y_train,y_pred))
print('Mean absolute error for traning :' , metrics.mean_squared_error(y_train,y_pred))
print('Root mean squared error for training :' , np.sqrt(metrics.mean_squared_error(y_train,y_pred)))
print()
y_pred=regressor.predict(x_test)
print('r2 score for testing :' ,r2_score(y_test,y_pred))
print('Mean absolute error for testing:', metrics.mean_absolute_error(y_test,y_pred))
print('Mean squared testing:', metrics.mean_squared_error(y_test,y_pred))
print('Root Mean Squared error for testing:' , np.sqrt(metrics.mean_squared_error(y_test,y_pred)))      
```

```python
#repeated random test_train splits croess validation
from sklearn.model_selection import ShuffleSplit
model=RandomForestRegressor()
ssplit=ShuffleSplit(n_splits=10,test_size=0.28)
from sklearn.model_selection import cross_val_score
results= cross_val_score(model,x,y,cv=ssplit)
print(results)
print(np.mean(results))
```
